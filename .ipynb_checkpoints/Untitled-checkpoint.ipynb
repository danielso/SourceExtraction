```python
%matplotlib inline
cmap='gnuplot'
import numpy as np
from pylab import load
import matplotlib.pyplot as plt
#from matplotlib.backends.backend_pdf import PdfPages
from AuxilaryFunctions import GetFileName  
from scipy.ndimage.filters import gaussian_filter
from Demo import GetDefaultParams

params,params_dict=GetDefaultParams()
last_rep=params.repeats
  
for rep in range(last_rep): 
    resultsName=GetFileName(params_dict,rep)
    try:
        results=load('NMF_Results/'+resultsName)
    except IOError:
        if rep==0:
            print 'results file not found!!'              
        else:
            break     
    S=results['shapes']
    A=results['activity']
    if rep>params.Background_num:
        adaptBias=False
    else:
        adaptBias=True
    L=len(A)-adaptBias 
    if L==0: #stop if we find files with zero components
        break
    if rep==0:
        shape_dims=S[0].shape
        D=len(shape_dims)
    shapes=S[:-adaptBias]
    activity=A[:-adaptBias]
    if rep==0:
        shapes=S
        activity=A
    else:
        shapes=np.append(shapes,S,axis=0)
        activity=np.append(activity,A,axis=0)  

L=len(shapes)
print L
```

```python
from scipy.signal import welch
width = 17
height = 12
plt.figure(figsize=(width, height*L))

for ll in range(L):
    image = shapes[ll]

    Y=image[image>0].ravel()
    L = len(Y)
    ff, psd_Y = welch(Y, nperseg=round(L / 8))
    #plt.plot(ff,psd_Y)
    sn = np.sqrt(np.mean(psd_Y[ff > .3] / 2))
#    print sn

    image2=(image>(sn))*image
    plt.subplot(L,3,1+ll)
    plt.imshow(image, cmap=cmap)
    plt.subplot(L,3,2+ll)
    plt.imshow(image2, cmap=cmap)
    plt.subplot(L,3,3+ll)
    plt.imshow(ndimage.median_filter(image2,3), cmap=cmap)
```

```python
per=90 #percentile threshold
for ll in range(L):
    temp0=shapes[ll]
    ind0=temp0>0
    mi=0
    ma=np.percentile(temp0[ind0],per)
    temp1=0*temp0
    temp1[ind0]=temp0[ind0]>np.percentile(temp0[ind0],per)
    sigma=1
    temp2=gaussian_filter(temp1,[sigma]*D)
    temp3=temp2>0.5/(np.sqrt(2*np.pi)*sigma)**D
    temp4=temp3-temp1

    from scipy.ndimage.measurements import label
    CC,num_CC=label(temp4)
    sz=0
    ind_best=0
    for nn in range(num_CC):
        current_sz=np.count_nonzero(CC[CC==(nn+1)])
        if current_sz>sz:
            ind_best=nn+1
            sz=current_sz
    temp4[CC!=ind_best]=0

    width = 17
    height = 12*L
    plt.figure(figsize=(width, height))

    b=3
    plt.subplot(L,b,b*ll+1)
    plt.imshow(temp0,vmin=mi,vmax=ma,cmap=cmap)
    plt.subplot(L,b,b*ll+2)
    plt.imshow(temp1,cmap=cmap)
    string='quality = ' + '%.2f' % (np.sum(temp4)/np.sum(temp2))
    plt.title(string)
    plt.subplot(L,b,b*ll+3)
    plt.imshow(temp4,cmap=cmap)

    
```

```python
#%% ###### Plot Individual neurons' activities
#%% plotting params
ComponentsInFig=20

left  = 0.05 # the left side of the subplots of the figure
right = 0.95   # the right side of the subplots of the figure
bottom = 0.05   # the bottom of the subplots of the figure
top = 0.95      # the top of the subplots of the figure
wspace = 0.1   # the amount of width reserved for blank space between subplots
hspace = 0.12  # the amount of height reserved for white space between subplots        
    
index=0 #component display index
sz=np.min([ComponentsInFig,L])
a=np.ceil(np.sqrt(sz))  
b=np.ceil(sz/a)  


for ii in range(L):
    if index==0:
        fig0=plt.figure(figsize=(11,18))
    ax = plt.subplot(a,b,index+1)
    pl=plt.plot(activity[ii])
    plt.setp(ax, xticks=[],yticks=[0])
    plt.setp(pl,'linewidth','2.0')
    # component number
    ax.text(0.02, 0.8, str(ii),
        verticalalignment='bottom', horizontalalignment='left',
        transform=ax.transAxes,
        color='black',weight='bold', fontsize=13)
    index+=1   
    if (ii%ComponentsInFig==(ComponentsInFig-1)) or ii==L-1: 
        plt.subplots_adjust(left, bottom, right, top, wspace, hspace)
        index=0
```
